# Removing Outliers

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5))
sns.boxplot(df.Dexa_Freq_During_Rx,ax = axes[0])
sns.distplot(df.Dexa_Freq_During_Rx,bins = 15, kde = False, ax = axes[1])
```

```python
sort=sorted(df['Dexa_Freq_During_Rx'])
q1=np.percentile(sort,25)
q3=np.percentile(sort,75)
IQR=q3-q1
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
```

I will removing the upper values with upper values and lower values also removing with lower bound

```python
h_out=np.where(df['Dexa_Freq_During_Rx']>upr_bound)
for i in range(len(h_out)):
  df['Dexa_Freq_During_Rx'][h_out[i]]=upr_bound

l_out=np.where(df['Dexa_Freq_During_Rx']<lwr_bound)
for i in range(len(h_out)):
  df['Dexa_Freq_During_Rx'][l_out[i]]=lwr_bound
```

```python
sns.boxplot(df['Dexa_Freq_During_Rx'])
```

# Removing Outliers from No.of Counts

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5))
sns.boxplot(df.Count_Of_Risks,ax = axes[0])
sns.distplot(df.Count_Of_Risks,bins = 15, kde = False, ax = axes[1])
```

```python
sort=sorted(df['Count_Of_Risks'])
q1=np.percentile(sort,25)
q3=np.percentile(sort,75)
IQR=q3-q1
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
```

```python
h_out=np.where(df['Count_Of_Risks']>upr_bound)
for i in range(len(h_out)):
  df['Count_Of_Risks'][h_out[i]]=upr_bound

l_out=np.where(df['Count_Of_Risks']<lwr_bound)
for i in range(len(h_out)):
  df['Count_Of_Risks'][l_out[i]]=lwr_bound
```

```python
sns.boxplot(df['Count_Of_Risks'])
```

# Features Corelation with Target Variable

```python
#Correlation of features to the terget
corr = df[df.columns[1:]].corr()['Persistency_Flag'][:]
corr_sorted = corr.sort_values(ascending=False)

#Let's see top 10 most correlated features 
corr_sorted[:]
```

The above correlation measures pointed out that the features with negative scores are not powerful in predicting target variable. So we have decided to take the features which are greater the value of 0.025.

As we see from the chart that the above mentioned features are the most relevant features for predicting the Persistency_flag.
